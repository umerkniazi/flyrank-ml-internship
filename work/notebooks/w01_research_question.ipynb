## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
from pathlib import Path
import pandas as pd

repo_root = Path.cwd().parents[1]
data_path = repo_root / "data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(data_path)

print("Loaded from:", data_path)
print("Total pages:", len(df))

print(
    "Declining pages:",
    (df["trend_direction"].str.lower() == "down").sum()
)

print(
    "Median impressions in 90 days:",
    df["impressions_90d"].median()
)

print(
    "Median days since last update:",
    df["days_since_last_update"].median()
)
```

# What these numbers suggest

The dataset contains a large number of pages with different levels of visibility, freshness, and performance trends.

The variation in impressions and update history suggests that pages have different levels of opportunity and risk.

These signals make content prioritization a useful problem to explore with data and ML.
